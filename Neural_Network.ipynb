## Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning.

### Answer:

#### What is Deep Learning?

Deep Learning is a subset of Machine Learning that uses Artificial Neural Networks with multiple hidden layers to learn patterns from large amounts of data automatically.

#### Evolution of Deep Learning

*   **1940s–1950s:** Introduction of artificial neurons by McCulloch and Pitts.
*   **1958:** Frank Rosenblatt developed the Perceptron.
*   **1980s:** Backpropagation algorithm enabled training of multi-layer neural networks.
*   **2006:** Geoffrey Hinton popularized Deep Learning techniques.
*   **2012-Present:** Growth of GPUs, large datasets, and improved algorithms led to breakthroughs in image recognition, speech processing, and natural language processing.

#### Difference Between Deep Learning and Traditional Machine Learning

| Traditional Machine Learning      | Deep Learning                       |
| :-------------------------------- | :---------------------------------- |
| Requires manual feature engineering | Learns features automatically       |
| Works well with small datasets    | Requires large datasets             |
| Faster training                   | Computationally expensive           |
| Suitable for structured data      | Suitable for structured and unstructured data |
| Examples: Decision Trees, SVM     | Examples: CNN, RNN, ANN             |

#### Conclusion

Deep Learning is a powerful advancement of Machine Learning capable of automatically extracting complex features and solving highly sophisticated problems.

## Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?

### Answer:

#### Perceptron Architecture

A Perceptron is the simplest neural network consisting of:

*   Input Layer
*   Weights
*   Summation Function
*   Activation Function
*   Output Layer

#### Working of Perceptron

The perceptron calculates:

$y = f(\sum w_i x_i + b)$

Where:

*   $x_i$ = Inputs
*   $w_i$ = Weights
*   $b$ = Bias
*   $f$ = Activation Function

#### Steps

1.  Receive input values.
2.  Multiply inputs by weights.
3.  Add bias.
4.  Apply activation function.
5.  Generate output.

#### Limitations

*   Solves only linearly separable problems.
*   Cannot solve XOR gate.
*   Limited learning capability.
*   Single-layer architecture.

#### Conclusion

Perceptrons form the foundation of neural networks but have limitations that are overcome by multilayer neural networks.

## Question 3: Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

### Answer:

#### Purpose of Activation Function

Activation functions introduce non-linearity into neural networks, enabling them to learn complex patterns.

#### Comparison

##### Sigmoid Function

*   **Formula:** $\sigma(x) = \frac{1}{1 + e^{-x}}$
*   **Range:** (0,1)
*   **Advantages:**
    *   Useful for binary classification.
*   **Disadvantages:**
    *   Suffers from vanishing gradient problem.

##### ReLU Function

*   **Formula:** $f(x) = \text{max}(0, x)$
*   **Range:** [0,∞)
*   **Advantages:**
    *   Computationally efficient.
    *   Reduces vanishing gradient problem.
*   **Disadvantages:**
    *   Dying ReLU problem.

##### Tanh Function

*   **Formula:** $\text{tanh}(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$
*   **Range:** (-1,1)
*   **Advantages:**
    *   Zero-centered output.
    *   Better gradient flow than Sigmoid.
*   **Disadvantages:**
    *   Still suffers from vanishing gradients.

#### Comparison Table

| Function | Range   | Advantage        | Disadvantage        |
| :------- | :------ | :--------------- | :------------------ |
| Sigmoid  | (0,1)   | Binary output    | Vanishing gradient  |
| ReLU     | [0,∞)   | Fast computation | Dead neurons        |
| Tanh     | (-1,1)  | Zero centered    | Vanishing gradient  |

#### Conclusion

ReLU is generally preferred in hidden layers due to better performance and faster convergence.

## Question 4: What is the difference between Loss function and Cost function in neural networks? Provide examples.

### Answer:

#### Loss Function

Loss measures error for a single training example.

**Example:**

$Loss = (y - \hat{y})^2$

#### Cost Function

Cost is the average loss over the entire dataset.

$Cost = \frac{1}{N} \sum Loss$

#### Differences

| Loss Function     | Cost Function         |
| :---------------- | :-------------------- |
| Error of one sample | Average error of all samples |
| Local measurement | Global measurement    |
| Used during computation | Used for optimization |

#### Examples

*   Mean Squared Error (MSE)
*   Binary Cross Entropy
*   Categorical Cross Entropy

#### Conclusion

Loss evaluates individual predictions, while cost evaluates overall model performance.

## Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.

### Answer:

#### Role of Optimizers

Optimizers update weights and biases to minimize the loss function.

#### Gradient Descent

*   Uses full dataset.
*   Simple and stable.
*   Slower convergence.

#### Adam

*   Combines Momentum and RMSprop.
*   Fast convergence.
*   Widely used.

#### RMSprop

*   Uses adaptive learning rates.
*   Effective for non-stationary problems.

#### Comparison

| Optimizer       | Speed  | Memory | Usage       |
| :-------------- | :----- | :----- | :---------- |
| Gradient Descent | Slow   | Low    | Basic tasks |
| RMSprop         | Medium | Medium | RNNs        |
| Adam            | Fast   | Medium | Most DL tasks |

#### Conclusion

Adam is generally preferred because it converges faster and performs well across many applications.

## Question 6: Single-Layer Perceptron Using NumPy for AND Gate

### Answer

#### Python Code

```python
import numpy as np

X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,0,0,1])

weights = np.array([1,1])
bias = -1.5

def step_function(x):
    return 1 if x >= 0 else 0

for x in X:
    net = np.dot(x, weights) + bias
    output = step_function(net)
    print(f"Input: {x}, Output: {output}")
```

## Question 7: Implement and Visualize Sigmoid, ReLU, and Tanh Activation Functions

### Answer

#### Python Code

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)

sigmoid = 1/(1+np.exp(-x))
relu = np.maximum(0,x)
tanh = np.tanh(x)

plt.figure(figsize=(10,6))

plt.plot(x,sigmoid,label='Sigmoid')
plt.plot(x,relu,label='ReLU')
plt.plot(x,tanh,label='Tanh')

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.grid()

plt.show()
```

## Question 8: Keras Neural Network on MNIST Dataset

### Answer


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train = X_train/255.0
X_test = X_test/255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_train,
    y_train,
    epochs=5,
    validation_split=0.2
)

loss, accuracy = model.evaluate(X_test, y_test)

print("Test Accuracy:", accuracy)
```

## Question 9: Fashion MNIST Loss and Accuracy Curves

### Answer


```python
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

(X_train,y_train),(X_test,y_test)=fashion_mnist.load_data()

X_train = X_train/255.0
X_test = X_test/255.0

model = Sequential([
    Flatten(input_shape=(28,28)),
    Dense(128,activation='relu'),
    Dense(10,activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_train,
    y_train,
    epochs=10,
    validation_split=0.2
)

plt.figure(figsize=(10,5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.legend()
plt.show()

plt.figure(figsize=(10,5))
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.legend()
plt.show()
```

#### Interpretation

*   Training loss decreases steadily.
*   Accuracy increases with epochs.
*   Small gap between training and validation curves indicates good generalization.
*   Large gaps would indicate overfitting.

## Question 10: Fraud Detection Using Deep Learning

### Answer

#### Model Design

A Multilayer Neural Network (MLP) is suitable because the dataset contains structured numerical and categorical features.

**Architecture:**

Input Layer
↓
Dense(64, ReLU)
↓
Dense(32, ReLU)
↓
Dense(1, Sigmoid)

#### Activation Function

*   **Hidden Layers:** ReLU
*   **Output Layer:** Sigmoid

**Reason:**

*   ReLU speeds up training.
*   Sigmoid provides probability for binary classification.

#### Loss Function

**Binary Cross Entropy**

`loss='binary_crossentropy'`

**Reason:**

*   Best suited for binary classification.

#### Handling Class Imbalance

*   Use Class Weights
*   Oversampling (SMOTE)
*   Precision, Recall, F1-score, ROC-AUC metrics

#### Optimizer

**Adam Optimizer**

**Reason:**

*   Fast convergence.
*   Adaptive learning rates.

#### Preventing Overfitting

*   Dropout Layers
*   Early Stopping
*   L2 Regularization

#### Python Code

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

model = Sequential([
    Dense(64, activation='relu', input_shape=(20,)),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

print(model.summary())
```
